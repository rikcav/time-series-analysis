```python
import pandas as pd
```

```python
df = pd.read_csv('./fb.csv', parse_dates=['Date'], index_col='Date')
df
```

```python
df.shift(1) # Shift prices by one day
```

```python
df.shift(-1)
```

```python
df['Price'].shift(2)
```

```python
df['Prev Day Price'] = df['Price'].shift(1)
df
```

```python
df['1 Day Change'] = df['Price'] - df['Prev Day Price']
df
```

```python
df['5 Day % Return'] = (df['Price'] - df['Price'].shift(5)) * 100 / df['Price'].shift(5)
df
```

```python
# Shift dates
df = df[['Price']]
df
```

```python
df.index # The frequency is None
```

```python
idx = pd.date_range(start='2017-08-15', periods=10, freq='B')
idx
```

```python
df.index = idx
df.index
```
